```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
x
```

```python
plt.plot(x, y)
plt.xlabel("Numbers")
plt.ylabel("Sin")
plt.title("Numbers and their sin")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label="sin(x)")
ax.plot(x, y2, label="cos(x)")

ax.set_xlabel("Numbers")
ax.set_ylabel("Sin(x)/Cos(x)")
ax.set_title("Cos and Sin of X")

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, label="sin(x)")
ax.plot(x, y2, label="cos(x)")

ax.set_xlabel("Numbers")
ax.set_ylabel("Sin(x)/Cos(x)")
ax.set_title("Cos and Sin of X")

ax.legend()
plt.savefig("sin_cos.pdf")
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, color="red", linestyle="solid", marker="o", label="sin(x)")
plt.plot(x, y2, color="blue", linestyle=":", marker="s", label="cos(x)")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)

plt.xticks(np.arange(0, 11, step=1), color="red", rotation=45)
plt.yticks(np.arange(-1, 1.1, step=0.5))

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.text(x=5, y=0, s="Text at (0,5)")
plt.annotate('Local max', xy=(1.57, 1), xytext=(2, 0.7),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
plt.plot(x,y)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,8), sharex=True)
ax1.plot(x, y1)
ax1.set_title("Sin(x)")
ax1.set_ylabel("Sin")

ax2.plot(x, y2)
ax2.set_title("Cos(x)")
ax2.set_ylabel("Cos")

ax2.set_xlabel("Number")
plt.show()
```
